#####  Calcule o salário bruto (horas * salario por hora)

```python
n1 = float(input('Qual o valor de sua hora de trabalho: '));
n2 = float(input('quantas horas voce trabalha por mes: '));

salario = n1 * n2;

print('Seu salario bruto por mes é R${}'.format(salario));
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
n1 = float(input('Qual o valor de sua hora de trabalho: '));
n2 = float(input('quantas horas voce trabalha por mes: '));

bruto = n1 * n2;
desconto11Ir = (bruto)*0.11;

print('Salario bruto é R${}'.format(bruto));
print('Desconto IR sobre salario bruto é R${}'.format(desconto11Ir));
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
n1 = float(input('Qual o valor de sua hora de trabalho: '));
n2 = float(input('quantas horas voce trabalha por mes: '));

bruto = n1 * n2;
descontoInss = (bruto)*0.08;

print('Salario bruto é R${}'.format(bruto));
print('Desconto IR sobre salario bruto é R${}'.format(descontoInss));
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
n1 = float(input('Qual o valor de sua hora de trabalho: '));
n2 = float(input('quantas horas voce trabalha por mes: '));

bruto = n1 * n2;
descontoSindicato = (bruto)*0.05;

print('Salario bruto é R${}'.format(bruto));
print('Desconto IR sobre salario bruto é R${}'.format(descontoSindicato));
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
n1 = float(input('Qual o valor de sua hora de trabalho: '));
n2 = float(input('quantas horas voce trabalha por mes: '));

bruto = n1 * n2;

descontoSindicato = (bruto)*0.05;
desconto11Ir = (bruto)*0.11;
descontoInss = (bruto)*0.08;

descontosTotal = descontoSindicato + desconto11Ir + descontoInss;

liquido = bruto - descontosTotal;

print('Salario bruto é R${}'.format(bruto));
print('Salario liquido é R${}'.format(liquido));
print('Desconto impostos sobre salario bruto é R${}'.format(descontosTotal));
```
